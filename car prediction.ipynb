```python
x = df[['Name', 'Company', 'Year', 'Kms_driven', 'Fuel_type']]
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe_arr = ohe.fit(x[['Name','Company','Fuel_type']])
ohe_arr
```

```python
ohe.categories_
```

```python
#transform the columns using make_column_transformer
column_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_),['Name','Company','Fuel_type']),
                                    remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
#It automatically go to the OneHotEncoder and transform each columns and pass to the LinearRegression for prediction
pipe = make_pipeline(column_trans, lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
score = []
for i in range(1000):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.1, random_state = i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans, lr)
    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    score.append(r2_score(y_test, y_pred))
```

```python
np.argmax(score)
```

```python
score[np.argmax(score)]
```
